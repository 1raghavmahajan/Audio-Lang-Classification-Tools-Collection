```python
# converting mp3 to wav
import os
source = './data/'
classes = ['english','hindi','urdu','telugu']
for Class in classes:
    path = source + Class 
    new_path = source + Class + '_wav'
    files = os.listdir(path)
    for file in files:
         !ffmpeg -i '{path}/{file}' -vn -acodec pcm_s16le -ac 1 -ar 22050 -f wav '{new_path}/{file[:-4]}.wav'
```
